```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/superstore_dataset2011-2015 (1).csv',encoding='latin-1',thousands=',',delimiter=',')
```

```python
df.head()
```

```python
# Chuyển đổi kiểu dữ liệu của thuộc tính "Sales" và "Profit" từ kiểu Object về kiểu float
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

```python
#Xóa giá trị null trong các trường
df.dropna(subset=['Sales', 'Quantity', 'Discount', 'Profit', 'OrderPriority'], inplace=True)
```

```python
discount_dtype = df['Profit'].dtype
df['Profit'] = df['Profit'].abs()
print(discount_dtype)
```

```python
features = df[['Sales', 'Quantity','Discount','Profit']]
labels = df['OrderPriority']
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Khởi tạo và huấn luyện mô hình Decision Tree
model = tree.DecisionTreeClassifier(max_depth=2)
model.fit(X_train, y_train)
```

```python
# Dự đoán nhãn cho dữ liệu kiểm tra
y_pred = model.predict(X_test)
```

```python
# Đánh giá độ chính xác của mô hình
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Visualize cây quyết định
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15, 8))
tree.plot_tree(model, feature_names=['Sales', 'Quantity', 'Discount', 'Profit'], class_names=df['OrderPriority'].unique(), filled=True, ax=ax)
plt.show()
```
